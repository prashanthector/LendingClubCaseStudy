```python
# For annual income let's plot the graph again
bins = [0, 10000, 50000, 100000, 500000, 1000000, 5000000, 6000000]
plt.figure(figsize=(10, 6))
sns.histplot(lending_df['annual_inc'], bins=bins, kde=False, color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
lending_df['annual_inc'].value_counts().sort_values(ascending=False)
```

```python
# Derive new column name annual_inc_bracket which will help us to analyze annual income amount in range
def annual_inc_bckt(column):
    if column <= 40000:
        return '0 - 40k'
    elif (column >40000) and (column <= 50000):
        return '40k - 50k'    
    elif (column >50000) and (column <= 60000):
        return '50k to 60k'
    elif (column >60000) and (column <= 70000):
        return '60k to 70k'
    elif (column >70000) and (column <= 80000):
        return '70k to 80k'
    else:
        return '80k - above'

lending_df['annual_inc_bckt'] = lending_df.apply(lambda x: annual_inc_bckt(x['annual_inc']), axis = 1)
```
